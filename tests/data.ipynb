```python
import random
import numpy as np

# sizes 1Kb 10Kb, 100Kb, 1Mb, 10Mb
sizes = ["1Kb", "10Kb", "100Kb", "1Mb", "10Mb"]
bandwidth_replication = np.array([2.98, 6.2, 7.23, 9.8, 12.11]) 
bandwidth_std = np.array([2.98, 6.2, 7.23, 9.8, 12.11])

time_replication = np.array([2.98, 6.2, 7.23, 9.8, 12.11])
time_std = np.array([2.98, 6.2, 7.23, 9.8, 12.11])

```

```python
import numpy as np
import matplotlib.pyplot as plt



# Plot the first set of bars with error bars
plt.bar(sizes, bandwidth_replication, yerr=bandwidth_std, color='skyblue', capsize=5)


# Add labels, title, and ticks
plt.xlabel('File Size')
plt.ylabel('Bandwidth (B/s)')
plt.title('Bandwidth Overhead for Replicating a File')

# Show the plot
plt.show()
```

```python
plt.bar(sizes, time_replication, yerr=time_std, color='skyblue', capsize=5)


# Add labels, title, and ticks
plt.xlabel('File Size')
plt.ylabel('Time (s)')
plt.title('Time Overhead for Replicating a File')

# Show the plot
plt.show()
```

```python
# Cache Performance

cache_size = np.array([0, 0.1, 0.2, 0.5, 1])
read_latency_zipfian = [100, 10, 20, 30, 40]
read_latency_uniform = [100, 10, 20, 30, 40]

std_zipfian = np.array([1, 2.98, 6.2, 7.23, 9.8])
std_uniform = np.array([1, 2.98, 6.2, 7.23, 9.8])

mixed_latency_zipfian = [100, 10, 20, 30, 40]
mixed_latency_uniform = [100, 10, 20, 30, 40]
std_mixed_zipfian = np.array([1, 2.98, 6.2, 7.23, 9.8])
std_mixed_uniform = np.array([1, 2.98, 6.2, 7.23, 9.8])

categories = ["Zapfian Distribution", "Uniform Distribution", "Mixed Load Zapfian", "Mixed Load Uniform Distribution"]
values = [read_latency_zipfian, read_latency_uniform, mixed_latency_zipfian, mixed_latency_uniform]
std = [std_zipfian, std_uniform, std_mixed_zipfian, std_mixed_uniform]
```

```python
bar_width = 0.2
x = np.arange(len(cache_size))  # Category positions

# Create the plot
fig, ax = plt.subplots()

# Plot each group of bars
for i in range(4):  # Loop over each set of bars within a category
    # Adjust position for each set of bars
    ax.bar(x + i * bar_width, 
           values[i], 
           yerr=std[i],
           width=bar_width, label=categories[i], capsize=5)

# Adding labels, title, and legend
ax.set_xlabel('Cache Size (%)')
ax.set_ylabel('Average Read Latency (ms)')
ax.set_title('Average Read Latency under different work loads')
ax.set_xticks(x + bar_width * 1.5)
ax.set_xticklabels(cache_size * 100)
ax.legend()

# Show plot
plt.show()
```
